```python
def cos_dist(a, b):
    return a@b/(np.linalg.norm(a)*np.linalg.norm(b))
```

```python
colmap_rots, esim_rots, R, t = find_colmap_esim_rotation_translation()
trans_rots = np.stack([e@R.T for e in colmap_rots])
```

```python
for idx in range(len(colmap_rots)):
    col_r, esim_r = colmap_rots[idx], esim_rots[idx]
    print([cos_dist(col_r[i], esim_r[i]) for i in range(3)])
```
